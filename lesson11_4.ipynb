```python
#讀取csv檔案為2維
import csv
with open('The way.csv',encoding='utf-8',newline='') as file:
    csv_reader = csv.reader(file)
    next(csv_reader)
    #print(type(csv_reader))
    csv_list = list(csv_reader)
    display(csv_list)
```

```python
for student in csv_list:
    if student[0] == '陳怡珊':
        print(student)
        break
```
