```python
import random
def 獲得分數() -> list:
    分數 = []
    for i in range(5):
        分數.append(random.randint(50, 100))
    return 分數

學生 = []
for i in range(50):
    分數 = 獲得分數()
    學生.append(分數)
學生
```
